```python
import pandas as pd
import requests
```

```python
url = "https://servicodados.ibge.gov.br/api/v3/agregados/6903/periodos/all/variaveis/all?localidades=N1[all]&classificacao=842[46618]"

response = requests.request("GET", url)

if response.status_code == 200:
    print('Consulta realizada com sucesso.')
else:
    print('Erro na consulta:', response.status_code)
```

```python
json = response.json()

json
```
